```python
r = np.random.randn(20,4)
df = pd.DataFrame(r, columns=['so1','so2','so3','so4'])
print(df.head())

print('Chon cot so1\n',df['so1'])
print('Chon nhieu dong\n', df[0:3])
print('Chon nhieu cot\n', df[['so1','so3']])
print('Chon nhieu dong, nhieu cot\n', df.iloc[:3, :2])
df = df.drop(columns=['so1','so4'])
print(df)
```

```python
df1 = pd.DataFrame({'Name':['John','Smith','Paul'],'Age': [25,30,50]},index = [0, 1, 2])
df2 = pd.DataFrame({'Name':['Adam','Smith'],'Age': [26,11]},index = [3,4])

df_concat = pd.concat([df1,df2])
print('Concate two DataFrame\n', df_concat)
print('Drop duplicatevalues in DataFrame\n', df_concat.drop_duplicates('Name'))
print('Sort values in DataFrame\n', df_concat.sort_values('Age'))

df_rename = df_concat.rename(columns={'Name':'Surname','Age':'Age_pupil'})
print('Rename columns of DataFrame\n', df_rename)
```
